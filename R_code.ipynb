```r
orthogroups_df_species <- read.table("Orthogroups.GeneCount_species", header=TRUE, stringsAsFactors=FALSE)

clean_name_species<- function(name) {
  name <- gsub("^['\"]+|['\"]+$", "", name)           # remove surrounding quotes
  name <- gsub("^GCA_[0-9.]+_|^GCF_[0-9.]+_", "", name) # remove GCA/GCF prefixes
  name <- gsub("\\..*$", "", name)                     # remove everything after first dot
  name <- gsub("_var_.*|_subsp_.*", "", name)         # remove variety/subspecies
  return(name)
}

colnames(orthogroups_df_species)[-1] <- make.unique(clean_name_species(colnames(orthogroups_df_species)[-1]), sep = "_dup")

cols_to_convert_species <- colnames(orthogroups_df_species)[2:(ncol(orthogroups_df_species)-1)]

orthogroups_df_species[cols_to_convert_species] <- lapply(orthogroups_df_species[cols_to_convert_species], function(x) as.integer(x > 0))

cols_for_plot_species <- colnames(orthogroups_df_species)[2:(ncol(orthogroups_df_species)-1)]  # exclude Orthogroup and Total

options(repr.plot.width = 16, repr.plot.height = 9)

upset_plot_species <- upset(orthogroups_df_species,
                    nsets = length(cols_for_plot_species),
                    sets = rev(cols_for_plot_species),
                    keep.order = TRUE,
                    order.by = "freq",
                    mb.ratio = c(0.3, 0.7),
                    number.angles = 0,
                    point.size = 2,
                    line.size = 0.5,
                    mainbar.y.label = "Number of Orthogroups",
                    sets.x.label = "Orthogroups per Species",
                    text.scale = c(1.3, 1.1, 1.0, 1.1, 1.1, 1.1),
                    main.bar.color = "#3B0F70",
                    sets.bar.color = "#FB6F4B",
                    matrix.color = "#2b82a7ff")

grid::grid.text("Orthogroup Distribution Across Species (Baseline - No Heat Proteins)",
                x = 0.5, y = 0.98,
                gp = grid::gpar(fontsize = 14, fontface = "bold"))

grid::pushViewport(grid::viewport(y = 0.45, height = 0.99))

print(upset_plot_species, newpage = FALSE)
```

```r
# PROCESS BASELINE RUN (Without Heat Proteins)

stats_transformed_baseline <- stats_perspecies_species %>%
  filter(X %in% c("Number of genes", "Number of genes in orthogroups", "Number of unassigned genes", "Percentage of genes in orthogroups", "Percentage of unassigned genes", "Number of genes in species-specific orthogroups")) %>%
  column_to_rownames("X") %>%
  t() %>%
  as.data.frame() %>%
  rownames_to_column("Species") %>%
  mutate(across(-Species, as.numeric))

cat("Dimensions of transformed data for Baseline:\n")
print(dim(stats_transformed_baseline))
cat("\nFirst few rows:\n")
print(head(stats_transformed_baseline, 3))

clean_name_baseline <- function(name) {
  name <- gsub("^['\"]+|['\"]+$", "", name)
  name <- gsub("^GCA_[0-9.]+_|^GCF_[0-9.]+_", "", name)
  name <- gsub("\\..*$", "", name)
  name <- gsub("_var_.*|_subsp_.*", "", name)
  return(name)
}

stats_transformed_clean_baseline <- stats_transformed_baseline %>%
  mutate(
    Species_clean = clean_name_baseline(Species),
    Total_Genes = `Number of genes`,
    Genes_in_Orthogroups = `Number of genes in orthogroups`,
    Unassigned_Genes = `Number of unassigned genes`,
    Species_Specific_Genes = `Number of genes in species-specific orthogroups`
  ) %>%
  mutate(
    # 1. Unassigned genes (as percentage of total genes)
    Pct_Unassigned = (Unassigned_Genes / Total_Genes) * 100,
    # 2. Species-specific genes (as percentage of genes in orthogroups)
    Pct_Species_Specific = (Species_Specific_Genes / Genes_in_Orthogroups) * 100,
    # 3. Shared genes (as percentage of genes in orthogroups)
    Shared_Genes = Genes_in_Orthogroups - Species_Specific_Genes,
    Pct_Shared = (Shared_Genes / Genes_in_Orthogroups) * 100
  ) %>%
  group_by(Species_clean) %>%
  slice(1) %>%
  ungroup()

stats_long_counts_baseline <- stats_transformed_clean_baseline %>%
  select(Species = Species_clean, Total_Genes,
         unassigned_genes = Unassigned_Genes,
         species_specific_genes = Species_Specific_Genes,
         shared_genes = Shared_Genes) %>%
  pivot_longer(cols = c(unassigned_genes, species_specific_genes, shared_genes),
               names_to = "Category",
               values_to = "Count")

stats_long_pct_baseline <- stats_transformed_clean_baseline %>%
  select(Species = Species_clean, Total_Genes,
         unassigned_genes = Pct_Unassigned,
         species_specific_genes = Pct_Species_Specific,
         shared_genes = Pct_Shared) %>%
  pivot_longer(cols = c(unassigned_genes, species_specific_genes, shared_genes),
               names_to = "Category",
               values_to = "Percentage")

stats_long_combined_baseline <- stats_long_counts_baseline %>%
  left_join(stats_long_pct_baseline, by = c("Species", "Category", "Total_Genes"))

species_order_baseline <- stats_transformed_clean_baseline %>%
  arrange(desc(Total_Genes)) %>%
  pull(Species_clean)

stats_long_ordered_baseline <- stats_long_combined_baseline %>%
  mutate(Species = factor(Species, levels = species_order_baseline))

total_labels_data_baseline <- stats_transformed_clean_baseline %>%
  mutate(Species = factor(Species_clean, levels = species_order_baseline)) %>%
  select(Species, Total_Genes)


stats_transformed_heat <- stats_perspecies_species_w_heat %>%
  filter(X %in% c("Number of genes", "Number of genes in orthogroups", "Number of unassigned genes", "Percentage of genes in orthogroups", "Percentage of unassigned genes", "Number of genes in species-specific orthogroups")) %>%
  column_to_rownames("X") %>%
  t() %>%
  as.data.frame() %>%
  rownames_to_column("Species") %>%
  mutate(across(-Species, as.numeric))

cat("\nDimensions of transformed data for With Heat Proteins:\n")
print(dim(stats_transformed_heat))
cat("\nFirst few rows:\n")
print(head(stats_transformed_heat, 3))

stats_transformed_clean_heat <- stats_transformed_heat %>%
  mutate(
    Species_clean = clean_name_baseline(Species),
    Total_Genes = `Number of genes`,
    Genes_in_Orthogroups = `Number of genes in orthogroups`,
    Unassigned_Genes = `Number of unassigned genes`,
    Species_Specific_Genes = `Number of genes in species-specific orthogroups`
  ) %>%
  mutate(
    Pct_Unassigned = (Unassigned_Genes / Total_Genes) * 100,
    Pct_Species_Specific = (Species_Specific_Genes / Genes_in_Orthogroups) * 100,
    Shared_Genes = Genes_in_Orthogroups - Species_Specific_Genes,
    Pct_Shared = (Shared_Genes / Genes_in_Orthogroups) * 100
  ) %>%
  group_by(Species_clean) %>%
  slice(1) %>%
  ungroup()

stats_long_counts_heat <- stats_transformed_clean_heat %>%
  select(Species = Species_clean, Total_Genes,
         unassigned_genes = Unassigned_Genes,
         species_specific_genes = Species_Specific_Genes,
         shared_genes = Shared_Genes) %>%
  pivot_longer(cols = c(unassigned_genes, species_specific_genes, shared_genes),
               names_to = "Category",
               values_to = "Count")

stats_long_pct_heat <- stats_transformed_clean_heat %>%
  select(Species = Species_clean, Total_Genes,
         unassigned_genes = Pct_Unassigned,
         species_specific_genes = Pct_Species_Specific,
         shared_genes = Pct_Shared) %>%
  pivot_longer(cols = c(unassigned_genes, species_specific_genes, shared_genes),
               names_to = "Category",
               values_to = "Percentage")

stats_long_combined_heat <- stats_long_counts_heat %>%
  left_join(stats_long_pct_heat, by = c("Species", "Category", "Total_Genes"))

species_order_heat <- stats_transformed_clean_heat %>%
  arrange(desc(Total_Genes)) %>%
  pull(Species_clean)

stats_long_ordered_heat <- stats_long_combined_heat %>%
  mutate(Species = factor(Species, levels = species_order_heat))

total_labels_data_heat <- stats_transformed_clean_heat %>%
  mutate(Species = factor(Species_clean, levels = species_order_heat)) %>%
  select(Species, Total_Genes)

# CREATE PLOTS FOR BOTH RUNS

options(repr.plot.width = 16, repr.plot.height = 9)

plot_baseline <- ggplot() +
  geom_col(data = stats_long_ordered_baseline,
           aes(x = Species, y = Count, fill = Category),
           position = "stack", width = 0.8) +
  geom_text(data = stats_long_ordered_baseline,
            aes(x = Species, y = Count, fill = Category,
                label = ifelse(Percentage > 2, paste0(round(Percentage, 1), "%"), "")),
            position = position_stack(vjust = 0.5),
            size = 2.5, color = "white", fontface = "bold",
            angle = 90, hjust = 0.5, vjust = 0.5) +
  geom_text(data = total_labels_data_baseline,
            aes(x = Species, y = Total_Genes, label = scales::comma(Total_Genes)),
            angle = 90, hjust = -0.1, vjust = 0.5,
            size = 2.8, fontface = "bold", color = "black") +
  scale_fill_manual(
    values = c("unassigned_genes" = "#FB6F4B",        # Orange-red
               "species_specific_genes" = "#B63679",  # Magenta
               "shared_genes" = "#3B0F70"),           # Dark purple
    labels = c("Shared Orthogroups", "Species-Specific Orthogroups", "Unassigned Genes")
  ) +
  labs(
    x = "Plant Species",
    y = "Number of Protein-Coding Genes",
    title = "Baseline: Three-Category Gene Distribution Across Plant Species",
    subtitle = "Total gene count shown above bars (vertical), percentage composition shown inside bars",
    fill = "Gene Category"
  ) +
  scale_y_continuous(
    labels = scales::comma,
    expand = expansion(mult = c(0.02, 0.15))
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 55, hjust = 1, size = 9, face = "italic"),
    axis.title.x = element_text(size = 12, face = "bold", margin = margin(t = 10)),
    axis.text.y = element_text(size = 9),
    axis.title.y = element_text(size = 12, face = "bold", margin = margin(r = 10)),
    plot.title = element_text(hjust = 0.5, face = "bold", size = 14),
    plot.subtitle = element_text(hjust = 0.5, size = 11, color = "gray50"),
    legend.position = "top",
    legend.title = element_text(face = "bold", size = 10),
    legend.text = element_text(size = 9),
    panel.grid.major = element_line(color = "gray90"),
    panel.grid.minor = element_blank()
  )

plot_heat <- ggplot() +
  geom_col(data = stats_long_ordered_heat,
           aes(x = Species, y = Count, fill = Category),
           position = "stack", width = 0.8) +
  geom_text(data = stats_long_ordered_heat,
            aes(x = Species, y = Count, fill = Category,
                label = ifelse(Percentage > 2, paste0(round(Percentage, 1), "%"), "")),
            position = position_stack(vjust = 0.5),
            size = 2.5, color = "white", fontface = "bold",
            angle = 90, hjust = 0.5, vjust = 0.5) +
  geom_text(data = total_labels_data_heat,
            aes(x = Species, y = Total_Genes, label = scales::comma(Total_Genes)),
            angle = 90, hjust = -0.1, vjust = 0.5,
            size = 2.8, fontface = "bold", color = "black") +
  scale_fill_manual(
    values = c("unassigned_genes" = "#FB6F4B",        # Orange-red
               "species_specific_genes" = "#B63679",  # Magenta
               "shared_genes" = "#3B0F70"),           # Dark purple
    labels = c("Shared Orthogroups", "Species-Specific Orthogroups", "Unassigned Genes")
  ) +
  labs(
    x = "Plant Species",
    y = "Number of Protein-Coding Genes",
    title = "With Heat Proteins: Three-Category Gene Distribution Across Plant Species",
    subtitle = "Total gene count shown above bars (vertical), percentage composition shown inside bars",
    fill = "Gene Category"
  ) +
  scale_y_continuous(
    labels = scales::comma,
    expand = expansion(mult = c(0.02, 0.15))
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 55, hjust = 1, size = 9, face = "italic"),
    axis.title.x = element_text(size = 12, face = "bold", margin = margin(t = 10)),
    axis.text.y = element_text(size = 9),
    axis.title.y = element_text(size = 12, face = "bold", margin = margin(r = 10)),
    plot.title = element_text(hjust = 0.5, face = "bold", size = 14),
    plot.subtitle = element_text(hjust = 0.5, size = 11, color = "gray50"),
    legend.position = "top",
    legend.title = element_text(face = "bold", size = 10),
    legend.text = element_text(size = 9),
    panel.grid.major = element_line(color = "gray90"),
    panel.grid.minor = element_blank()
  )

print(plot_baseline)
print(plot_heat)

cat("\n SUMMARY FOR BOTH RUNS:\n")
cat("=========================\n")
cat("Baseline - Number of species:", nrow(stats_transformed_clean_baseline), "\n")
cat("With Heat Proteins - Number of species:", nrow(stats_transformed_clean_heat), "\n")
cat("Common species analysis completed for both runs.\n")
```

```r
# PRINT IN READABLE TABLE FORMAT

print_species_plot_data <- function(df, run_name) {
  cat(paste0("\n ", run_name, " - GENE DISTRIBUTION BY SPECIES:\n"))
  cat(paste0(rep("=", 80), collapse = ""), "\n")

  species_list <- levels(df$Species)

  for(species in species_list) {
    species_data <- df %>% filter(Species == species)
    total_genes <- species_data$Total_Genes[1]

    cat(paste0("\n", species, " (Total: ", scales::comma(total_genes), " genes)\n"))
    cat(rep("-", nchar(species) + 15), "\n")

    for(row in 1:nrow(species_data)) {
      category <- case_when(
        species_data$Category[row] == "shared_genes" ~ "Shared Orthogroups",
        species_data$Category[row] == "species_specific_genes" ~ "Species-Specific Orthogroups",
        species_data$Category[row] == "unassigned_genes" ~ "Unassigned Genes"
      )
      count <- species_data$Count[row]
      percentage <- round(species_data$Percentage[row], 1)

      cat(sprintf("  • %-30s: %9s genes (%5.1f%%)\n",
                  category,
                  scales::comma(count),
                  percentage))
    }
  }
}

print_species_plot_data(stats_long_ordered_baseline, "BASELINE RUN")
print_species_plot_data(stats_long_ordered_heat, "WITH HEAT PROTEINS RUN")
```

```r
library(ape)
library(vegan)
library(ggplot2)
library(dplyr)
library(tidyr)

analyze_phylo_expression_correlation <- function(gene_count_df, species_tree_file) {

  expression_data <- gene_count_df[, -c(1:2)]  # Remove Orthogroup and Total columns
  rownames(expression_data) <- gene_count_df$Orthogroup

  species_cor <- cor(expression_data, method = "pearson", use = "complete.obs")
  print("Species correlation matrix:")
  print(species_cor)

  species_tree <- read.tree(species_tree_file)
  phylo_dist <- cophenetic(species_tree)

  common_species <- intersect(colnames(expression_data), rownames(phylo_dist))
  expression_data <- expression_data[, common_species]
  phylo_dist <- phylo_dist[common_species, common_species]
  species_cor <- cor(expression_data, method = "pearson", use = "complete.obs")

  expression_distance <- 1 - species_cor

  lower_expr <- expression_distance[lower.tri(expression_distance)]
  lower_phylo <- phylo_dist[lower.tri(phylo_dist)]

  cor_test <- cor.test(lower_phylo, lower_expr, method = "pearson")

  lm_fit <- lm(lower_expr ~ lower_phylo)
  lm_summary <- summary(lm_fit)

  results <- data.frame(
    PhylogeneticDistance = lower_phylo,
    ExpressionDistance = lower_expr
  )

  return(list(
    results = results,
    correlation = cor_test,
    regression = lm_summary,
    correlation_matrix = species_cor,
    phylogenetic_distance = phylo_dist
  ))
}

print("=== ANALYSIS FOR SPECIES DATASET ===")
analysis_species <- analyze_phylo_expression_correlation(gene_count_species, "/content/SpeciesTree_rooted_species")

print("=== ANALYSIS FOR SPECIES + HEAT DATASET ===")
analysis_species_heat <- analyze_phylo_expression_correlation(gene_count_species_w_heat, "/content/SpeciesTree_rooted_species_w_heat")

print("Species dataset results:")
print(paste("PCC =", round(analysis_species$correlation$estimate, 3)))
print(paste("P-value =", round(analysis_species$correlation$p.value, 5)))
print(paste("R-squared =", round(analysis_species$regression$r.squared, 3)))

print("Species + Heat dataset results:")
print(paste("PCC =", round(analysis_species_heat$correlation$estimate, 3)))
print(paste("P-value =", round(analysis_species_heat$correlation$p.value, 5)))
print(paste("R-squared =", round(analysis_species_heat$regression$r.squared, 3)))
```

```r
extract_species_names <- function(gene_count_df) {
  species_cols <- colnames(gene_count_df)
  species_cols <- species_cols[!species_cols %in% c("Orthogroup", "Total")]
  clean_names <- gsub("GCF_[0-9]+\\.\\d+_", "", species_cols)
  clean_names <- gsub("\\.ref$", "", clean_names)
  clean_names <- gsub("_", " ", clean_names)
  return(clean_names)
}

species_names <- extract_species_names(gene_count_species)
species_names_heat <- extract_species_names(gene_count_species_w_heat)

print("Species in main dataset:")
print(species_names)

print("Species in heat dataset:")
print(species_names_heat)
```

```r
create_species_tree <- function(species_list) {
  asteraceae <- species_list[grep("Lactuca|Daucus", species_list)]
  solanaceae <- species_list[grep("Solanum", species_list)]
  poaceae <- species_list[grep("Oryza|Zea|Hordeum", species_list)]
  rosaceae <- species_list[grep("Malus|Pyrus|Mangifera", species_list)]
  vitaceae <- species_list[grep("Vitis", species_list)]
  cucurbitaceae <- species_list[grep("Cucumis", species_list)]
  fabaceae <- species_list[grep("Pisum", species_list)]
  others <- setdiff(species_list, c(asteraceae, solanaceae, poaceae, rosaceae, vitaceae, cucurbitaceae, fabaceae))

  asteraceae_clade <- if(length(asteraceae) > 1) paste0("(", paste(asteraceae, collapse = ","), ")") else asteraceae
  poaceae_clade <- if(length(poaceae) > 1) paste0("(", paste(poaceae, collapse = ","), ")") else poaceae
  rosaceae_clade <- if(length(rosaceae) > 1) paste0("(", paste(rosaceae, collapse = ","), ")") else rosaceae

  tree_text <- paste0(
    "(((",
    paste(c(asteraceae_clade, solanaceae, cucurbitaceae), collapse = ","),
    "),(",
    paste(c(rosaceae_clade, vitaceae, fabaceae), collapse = ","),
    ")),",
    "(",
    paste(c(poaceae_clade, others), collapse = ","),
    "));"
  )

  tree <- read.tree(text = tree_text)
  return(tree)
}

species_tree <- create_species_tree(species_names)
species_tree_heat <- create_species_tree(species_names_heat)

print("Species tree for main dataset:")
print(species_tree)

print("Species tree for heat dataset:")
print(species_tree_heat)
```
